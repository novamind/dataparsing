```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import re
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}

url = 'https://www.azlyrics.com/m/mileycyrus.html'
# url = 'https://www.azlyrics.com/t/tylerthecreator.html'
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')

data = []
for s in soup.select('.listalbum-item'): #get list of songs
    
    song_page_link = s.a.get('href')
    if song_page_link.startswith('..'):
        song_page_link = 'https://www.azlyrics.com' + s.a.get('href').replace('..', '')
        
    print('Trying to access', song_page_link, end='')
    
    page_content = requests.get(song_page_link, headers=headers)
    page_soup = BeautifulSoup(page_content.content, 'html.parser')
    
    lyric = page_soup.select_one('.col-xs-12.col-lg-8.text-center').find('div', class_=None)
    song_detail = page_soup.select_one('.feat')
    
    album = page_soup.select_one('.songinalbum_title')
    
    album_details = None
    album_title = None
    if album != None and album.text != 'You May Also Like':
        album_details = album.text
        album_title = album.b.text
    song_detail_text = None
    if song_detail != None:
        song_detail_text = song_detail.text
        
    data.append({
        'album_info': album_details, 
        'album_title': album_title,
        'lyrics': lyric.text,
        'song_title': s.a.text,
        'song_detail': song_detail_text
    })
    print('... succesfully')
    time.sleep(15)
```

```python
df = pd.DataFrame(data)
print(f'Got {len(df)} lyrics')
display(df.sample(5))
```

```python
df.to_csv('data/lyrics_mileycyrus.csv')
```
